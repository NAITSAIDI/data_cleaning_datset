```python
#Function to fetch and reshape
```

```python
def get_n_melt_n_data(data_url, case_type):
    df= pd.read_csv(data_url)
    melted_df= df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'])
    melted_df.rename(columns= {'variable': 'Date','value': case_type}, inplace= True)
    return  melted_df
```

```python
# Fetch death or recovered 
```

```python
recovered_df=  get_n_melt_n_data(recovered_cases_url, 'Recovered')
death_df= get_n_melt_n_data( death_cases_url, 'Deaths' )
```
